## Importing Libraries and Dataset

```python
# import all modules
% matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the DataFrame
df = pd.read_csv('auto-mpg.csv')
df.tail()
```

```python
df.describe()
```

## 1. Categorical Plot Types

### `a. stripplot()` 

```python
# Create the stripplot
sns.stripplot(data=df,
         x='mpg',
         y='origin',
         jitter=True)

plt.show()
```

### b.  `swarmplot()`

```python
# Create and display a swarmplot with hue set to the color
sns.swarmplot(data=df,
         x='mpg',
         y='color',
         hue='origin')

plt.show()
```

### c. boxplot()

```python
# Create a boxplot
sns.boxplot(data=df,
         x='mpg',
         y='color')
plt.show()
plt.clf()
```

### d. violinplot()

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df, x='mpg', y='color', palette='husl')

plt.show()
plt.clf()
```

### e. boxenplot()  
       (lvplot() is beeing used as boxenplot() anymore.)

```python
# Create a lvplot with the Paired palette and the origin column as the hue
sns.boxenplot(data=df,
         x='mpg',
         y='color',
         palette='Paired',
         hue='origin')

plt.show()
plt.clf()
```

### f. countplot()

```python
# Show a countplot 
sns.countplot(data=df,
         y="color",
         hue="origin")
plt.show()
plt.clf()
```

### g. pointplot()

```python
# Create a pointplot and include the capsize in order to show bars on the confidence interval
sns.pointplot(data=df,
         y='hp',
         x='origin',
         capsize=.1)
plt.show()
plt.clf()
```

### h. barplot()

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='hp',
         x='color',
         hue='origin')
plt.show()
plt.clf()
```

## 2. Regression Plots

### a. regplot()

```python
# Display a regression plot for Tuition
sns.regplot(data=df,
         y='mpg',
         x="hp",
         marker='^',
         color='g')

plt.show()
plt.clf()
```

* Using "x_bins"

```python
sns.regplot(data=df,
         y='mpg',
         x="hp",
         marker='^',
         color='g',
         x_bins=5)

plt.show()
plt.clf()
```

* Using "order"

```python
sns.regplot(data=df,
         y='mpg',
         x="hp",
         marker='^',
         color='g',
         x_bins=5,
         order=2)

plt.show()
plt.clf()
```

* fit_reg parameter
   When you define fit_reg parameter as False, regressin line will get lost.

```python
# Create a scatter plot by disabling the regression line
sns.regplot(data=df,
            y='mpg',
            x="hp",
            fit_reg=False)

plt.show()
plt.clf()
```

### b. residplot()

```python
# Display the residual plot
sns.residplot(data=df,
          y='mpg',
          x="hp",
          color='g')

plt.show()
plt.clf()
```

## 3. Matrix Plots

### a. Creating heatmaps

A heatmap is a common matrix plot that can be used to graphically summarize the relationship between two variables.

```python
# Create a crosstab table of the data
pd_crosstab = pd.crosstab(df["mpg"], df["hp"])
print(pd_crosstab)
```

```python
# Plot a heatmap of the table
sns.heatmap(pd_crosstab)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```

### b. Customizing heatmaps
Seaborn supports several types of additional customizations to improve the output of a heatmap.

```python
sns.heatmap(pd_crosstab, cbar=False, cmap="Dark2", linewidths=.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

#Show the plot
plt.show()
plt.clf()
```
