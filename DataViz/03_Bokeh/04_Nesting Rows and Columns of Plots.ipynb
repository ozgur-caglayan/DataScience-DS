## Importing Libraries 

```python
import pandas as pd

from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.io import output_notebook   # To display the plot in jupyter inline


from bokeh.layouts import row
from bokeh.layouts import column
```

```python
output_notebook()
```

```python
# Read in the CSV file: df
df = pd.read_csv('auto-mpg.csv')
df.head()
```

```python
# Import row from bokeh.layouts
from bokeh.layouts import row

# Create the figure: p1
p1 = figure(x_axis_label='HP', y_axis_label='MPG')
p1.circle(df['hp'], df['mpg'], color=df['color'], size=10)

# Create the figure: p2
p2 = figure(x_axis_label='HP', y_axis_label='ACCEL')
p2.circle(df['hp'], df['accel'], color=df['color'], size=10)

# Create the figure: p1
p3 = figure(x_axis_label='HP', y_axis_label='WEIGHT')
p3.circle(df['hp'], df['weight'], color=df['color'], size=10)

# Make a column layout that will be used as the second row: row2
row2 = column([p1, p2], sizing_mode='scale_width')

# Make a row layout that includes the above column layout: layout
layout = row([p3, row2], sizing_mode='scale_width')

# Specify the name of the output file and show the result
output_file('nested.html')
show(layout)
```
