## Importing Libraries 

```python
import pandas as pd

from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.io import output_notebook   # To display the plot in jupyter inline

from bokeh.models.widgets import Panel
```

```python
output_notebook()
```

# Importing Dataset

```python
# Read in the CSV file: df
df = pd.read_csv('auto-mpg.csv')
df.head()
```

## Tabbed Layouts

Tabbed layouts can be created in Bokeh by placing plots or layouts in Panels.

### Creating Figures

```python
# Create the figure: p1
p1 = figure(x_axis_label='HP', y_axis_label='MPG')
p1.circle(df['hp'], df['mpg'], color=df['color'], size=10)

# Create the figure: p2
p2 = figure(x_axis_label='HP', y_axis_label='ACCEL')
p2.circle(df['hp'], df['accel'], color=df['color'], size=10)

# Create the figure: p1
p3 = figure(x_axis_label='HP', y_axis_label='WEIGHT')
p3.circle(df['hp'], df['weight'], color=df['color'], size=10)

p4 = figure(x_axis_label='HP', y_axis_label='WEIGHT')
p4.circle(df['hp'], df['weight'], color=df['color'], size=10)
```

### Creating Tabs

```python
# Import Panel from bokeh.models.widgets
from bokeh.models.widgets import Panel

# Create tab1 from plot p1: tab1
tab1 = Panel(child=p1, title='P1')

# Create tab2 from plot p2: tab2
tab2 = Panel(child=p2, title='P2')

# Create tab3 from plot p3: tab3
tab3 = Panel(child=p3, title='P3')

# Create tab4 from plot p4: tab4
tab4 = Panel(child=p4, title='P4')
```

### Displaying Tabbed Layouts

```python
# Import Tabs from bokeh.models.widgets
from bokeh.models.widgets import Tabs

# Create a Tabs layout: layout
layout = Tabs(tabs=[tab1, tab2, tab3, tab4])

# Specify the name of the output_file and show the result
output_file('tabs.html')
show(layout)
```
