## Importing Libraries 

```python
import pandas as pd

from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.io import output_notebook   # To display the plot in jupyter inline

from bokeh.layouts import column
```

## Importing Dataset

```python
# Read in the CSV file: df
df = pd.read_csv('auto-mpg.csv')
df.head()
```

# Creating columns of plots

In this exercise, you're going to use the column() function to create a single column layout of the two plots you created in the previous exercise.

```python
# Import row from bokeh.layouts
from bokeh.layouts import column

# Create the figure: p1
p1 = figure(x_axis_label='HP', y_axis_label='MPG')
p1.circle(df['hp'], df['mpg'], color=df['color'], size=10)

# Create the figure: p2
p2 = figure(x_axis_label='HP', y_axis_label='ACCEL')
p2.circle(df['hp'], df['accel'], color=df['color'], size=10)

# Put p1 and p2 into a horizontal row: layout
layout = column([p1, p2], sizing_mode='scale_width')

# Specify the name of the output file and show the result
output_file('column_of_plots.html')
show(layout)
```
